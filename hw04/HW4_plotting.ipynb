```python
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

google = pd.read_csv('google_data.txt',sep='\t')
yahoo = pd.read_csv('yahoo_data.txt',sep='\t')
ny = pd.read_csv('ny_temps.txt',sep='\t')

fig, ax1 = plt.subplots(figsize=(15,8))

p1,=ax1.plot(google['Modified Julian Date'],google['Stock Value'], color='green')
p2,=ax1.plot(yahoo['Modified Julian Date'],yahoo['Stock Value'], color='purple')
ax1.tick_params('y',colors='green')
ax1.set_ylabel('Value (Dollars)',color='purple')
ax1.set_xlabel('Date (MJD)')
ax1.set_title('New York Temperature, Google and Yahoo!',fontweight='bold')
ax1.yaxis.set_minor_locator(MultipleLocator(20))
ax1.xaxis.set_minor_locator(MultipleLocator(200))

ax2=ax1.twinx()
p3,=ax2.plot(ny['Modified Julian Date'],ny['Max Temperature'], 'b--')
ax2.set_ylabel('Temperature ('+u'\N{DEGREE SIGN}'+'F)',color='blue')
ax2.tick_params('y',colors='blue')
ax2.set_ylim(-150,100)
ax2.yaxis.set_minor_locator(MultipleLocator(10))


plt.legend((p2,p1,p3),('Yahoo! Stock Value','Google Stock Value', 'NY Mon. High Temp'), loc=[0.01,0.35]);
plt.show()
```
